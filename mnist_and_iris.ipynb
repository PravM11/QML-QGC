```python
# convert images from grayscale to black and white

first_img = train_X[0]
for i in range(len(first_img)):
  for j in range(len(first_img[i])):
    if first_img[i][j] <= 127:
      first_img[i][j] = 0
    else:
      first_img[i][j] = 255

first_img
```

```python
# get the average color of each block
def get_average(starting_a, starting_b, img):
  sum = 0
  for i in range(starting_a, starting_a + 8):
    for j in range(starting_b, starting_b + 8):
      sum += img[i][j]
  if sum >= 5000:
    return 1 # black
  else:
    return 0 # white
```
